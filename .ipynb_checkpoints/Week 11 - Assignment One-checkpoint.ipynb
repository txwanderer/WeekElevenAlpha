```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
import math

%pylab inline
```

```python
dataFrame = pd.read_csv("breast_cancer.csv")
```

```python
dataFrame.head()
```

```python
dataFrame = dataFrame.drop(['Unnamed: 0', 'id number'], axis=1)
```

```python
y = dataFrame.pop("malignant")
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataFrame, y, test_size=.2, random_state=42)
```

```python
n_estimators = [300,400,500]
max_features = ['auto', 'sqrt','log2']
min_samples_split = [3,5,7]

rfc = RandomForestClassifier(n_jobs=1)
estimator = GridSearchCV(rfc,
                         dict(n_estimators=n_estimators,
                              max_features=max_features,
                              min_samples_split=min_samples_split
                              ), cv=None, n_jobs=-1)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
bestimator = estimator.best_estimator_
```

```python
accuracy = accuracy_score(y_test, bestimator.predict(X_test))
print("Accuracy: ", accuracy)
```

```python
print(classification_report(y_test, bestimator.predict(X_test)))
```

```python
roc = roc_auc_score(y_test, bestimator.predict_proba(X_test)[:,1])
print("AUC Score: ", roc)
```

```python
from sklearn import cross_validation
scores = cross_validation.cross_val_score(bestimator, dataFrame, y, cv=10, n_jobs=-1)
```

```python
scores
```

```python
mean_score = scores.mean()
std_dev = scores.std()
std_error = scores.std() / math.sqrt(scores.shape[0])
ci =  2.262 * std_error
lower_bound = mean_score - ci
upper_bound = mean_score + ci

print("Score is %f +/-  %f" % (mean_score, ci))
print("95 percent probability that if this experiment were repeated over and over the average score would be between %f and %f" % (lower_bound, upper_bound))
```
